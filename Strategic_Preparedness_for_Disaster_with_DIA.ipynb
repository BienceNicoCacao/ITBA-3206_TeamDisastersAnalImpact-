## **Dataset of Disasters**


---



```python
url = 'https://raw.githubusercontent.com/BienceNicoCacao/ITBA-3206_TheEcoWarriors/master/philippines_2020-significant-events1.csv'
```

```python
df= pd.read_csv(url)
df.head(1621)
```

## **Data Cleanning**

---

##### **_Dropping unnecessary columns_**


```python
df= pd.DataFrame(df)
df.drop(['Evacuation (Cummulative)', 
         'Evacuation (Now)'], axis = 1, inplace=True)
display(df)


```

##### **_Check nulls in column_**

```python
df.isnull().sum()
```

##### **_Removing Duplicates_**

```python
dupes = df.duplicated()
print(dupes[dupes == True])

df.drop_duplicates(inplace=True)
df.head(400)
```

## **Data Visualization**

```python
import plotly.graph_objects as go

labels = ['Taal Volcano Eruption','Earthquake (6.6 mag)','TY MOLAVE (Quinta)']
values = [4500, 2500, 1053]

# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0.2])])
fig.show()
```
